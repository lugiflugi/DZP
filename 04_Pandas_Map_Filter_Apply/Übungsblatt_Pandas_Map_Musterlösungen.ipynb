***1.***
An wie vielen Tagen wurde die Temperatur zwischen 5 und 10 Grad an der Rosengartenstrasse gemessen?

Für das filtern des DataFrames gibt es unterschiedliche Ansätze. Verwende die unterschiedlichen Methoden um die Anzahl der Tage zu ermitteln und mache ein Zeitmessung.

- Boolesche Indizierung: Filtert Daten basierend auf einem booleschen Ausdruck, der auf einer Spalte oder einer Zeile basiert.
- `query()`: Filtert Daten basierend auf einem Ausdruck, der in einer speziellen Abfragesprache (ähnlich SQL) geschrieben ist.
- `.loc[]` und `.iloc[]` : Filtert Daten basierend auf den Achsenbezeichnungen (wie Spaltennamen oder Indexwerte)


```python
import pandas as pd

df = pd.read_csv("ugz_ogd_meteo_d1_2023.csv")
df.sample(5)
```

```python
df_filtered = df.query(
    "Standort == 'Zch_Rosengartenstrasse' & Wert > 5 & Wert < 10 & Parameter == 'T'"
)
len(df_filtered)
```

```python
%%timeit
df_filtered = df.query("Standort == 'Zch_Rosengartenstrasse' & Wert > 5 & Wert < 10 & Parameter == 'T'")
```

```python
df_filtered = df.loc[
    (df["Standort"] == "Zch_Rosengartenstrasse")
    & (df["Wert"] > 5)
    & (df["Wert"] < 10)
    & (df["Parameter"] == "T")
]
print(len(df_filtered))
```

```python
%%timeit 
df_filtered = df.loc[
    (df["Standort"] == "Zch_Rosengartenstrasse") & 
    (df["Wert"] > 5) & 
    (df["Wert"] < 10) & 
    (df["Parameter"] == "T")]
```

```python
df_filtered = df[
    (df["Standort"] == "Zch_Rosengartenstrasse")
    & (df["Wert"] > 5)
    & (df["Wert"] < 10)
    & (df["Parameter"] == "T")
]
len(df_filtered)
```

```python
%%timeit 
df_filtered = df[
                (df["Standort"] == "Zch_Rosengartenstrasse") & 
                (df["Wert"] > 5) & 
                (df["Wert"] < 10) & 
                (df["Parameter"] == "T")
                ]
```

1.1 Was sind die Vor- und Nachteile der unterschiedlichen Methoden?
1. Direkt mit booleschem Indexing (df[...])
- Vorteile:
  - Schnellste Methode (minimaler Overhead)
  - Klarer, wenn nur Zeilen gefiltert werden 
- Achtung:
  - Nur für Zeilenfilter, kein Zugriff auf Spalten
2. Mit .loc[]
   - Vorteile:
     - Sehr gut lesbar und robust
     - Ermöglicht auch gleichzeitigen Spaltenzugriff (z.B. df.loc[...,'Spalte'])
   - Etwas langsamer als boolesches Indexing, wegen zusätzlichem Funktionalität
3. Mit .query()
   - Vorteile:
     - Sehr leserlich, besonders bei vielen Bedingungen
     - Gut für interaktive Analysen
   - Langsamste Methode, da 
     - String muss geparsed und interpretiert werden
     - Kann bei Sonderzeichen oder komplexen Spaltennamen problematisch sein
  
1.2 Weshalb ist die .loc[] Methode robuster als die booleschem Indexing (df[...]) Methode?

1. Explizite Absicht:
   - .loc[rows] sagt klar: „Ich greife auf Zeilen mit bestimmten Labels oder Bedingungen zu.“
   - df[condition] funktioniert nur, wenn condition ein boolescher Vektor ist.
   - Wenn du versehentlich etwas anderes in df[...] reinschreibst (z. B. ein einzelner Spaltenname), bekommst du etwas ganz anderes zurück:
   - Ambiguität: df[...] kann Spaltenzugriff oder Filter sein. .loc[...] ist eindeutig.
  
2. Erweiterbarkeit mit Spaltenzugriff
   - .loc[rows, columns] (df.loc[(df["Parameter"] == "T"),["Wert", "Standort"]]) erlaubt später einfach die Auswahl von Spalten zusätzlich zur Filterung:
   - Mit df[condition] geht das nicht direkt. Man müsste danach nochmal df_filtered[["Wert", "Standort"]] schreiben.
   - 

3. Mehr Fehlerschutz bei komplexem Code
    - Bei verschachteltem Code ist .loc deutlicher und besser gegen Fehler abgesichert, weil die Absicht klar ist: „Ich will Zeilen mit diesen Bedingungen.“


Fazit: Für produktiven, stabilen Code ist .loc[] robuster und sicherer – auch wenn df[...] minimal schneller ist.

## Aufgabe 3 - BMI-Rechner für Fitnessmitglieder

***0.*** Lies die Daten aus der Datei "ko_fitness_memberlist.csv" in ein DataFrame ein.

***1.*** Verwende die map() bzw. apply()-Funktion um die Spalten bmi und bmi-meldung zu berechnen

```python
import pandas as pd


def bmi_meldung(bmi):

    return_val = "Unter 18.5: Untergewicht."
    if bmi > 40:
        return_val = "Über 40: Adipositas Grad III."
    elif bmi >= 35:
        return_val = "35 - 39.9: Adipositas Grad II."
    elif bmi >= 30:
        return_val = "30 - 34.9: Adipositas (Fettleibigkeit) Grad I."
    elif bmi >= 25:
        return_val = "25 - 29.9: Übergewicht/ Präadipositas."
    elif bmi >= 18.5:
        return_val = "18.5 - 24.9: Normalgewicht."
    return return_val


def bmi_rechner(koerper_gewicht, koerper_groesse):
    return koerper_gewicht / koerper_groesse**2
```

```python
df_member_list = pd.read_csv("ko_fitness_memberlist.csv", delimiter=";")


df_member_list["BMI"] = df_member_list.apply(
    lambda row: bmi_rechner(row["Gewicht"], row["Grösse"]), axis=1
)

df_member_list["BMI Meldung"] = df_member_list["BMI"].map(bmi_meldung)

df_member_list
```

***2.*** Finde heraus, welche Mitglieder übergewichtig sind und wieviele Männer bzw. Frauen untergewichtig sind.

```python
underweight_female = df_member_list.loc[
    (df_member_list["Geschlecht"] == "w") & (df_member_list["BMI"] < 18.5)
]
```

***3.*** Berechne die Durschnittsgrösse, -gewicht und BMI aller 

- männlichen Mitglieder, welche grösser als 1.80m sind.
- weiblichen Mitglieder, die untergewichtig sind
- übergewichtigen Mitglieder

Versuche die Aufgabe mit einer Methode-Chaining zu lösen.

```python
# Berechne die Durschnittsgrösse, -gewicht und BMI aller männlichen Mitglieder, welche grösser als 1.80m sind.
average_over_180_male = (
    df_member_list.loc[
        (df_member_list["Geschlecht"] == "m") & (df_member_list["Grösse"] > 1.8)
    ]
    .filter(["Grösse", "Gewicht", "BMI"])
    .mean()
)
average_over_180_male


average_over_180_male = df_member_list.loc[
    (df_member_list["Geschlecht"] == "m") & (df_member_list["Grösse"] > 1.8),
    ["Grösse", "Gewicht", "BMI"],
].mean()

average_over_180_male
```

```python
# Berechne die Durschnittsgrösse, -gewicht und BMI aller  weiblichen Mitglieder, welche die Untergewichtig sind
average_underweighted_female = df_member_list.loc[
    (df_member_list["Geschlecht"] == "w") & (df_member_list["BMI"] < 18.5),
    ["Grösse", "Gewicht", "BMI"],
].mean()
average_underweighted_female
```

```python
# Berechne die Durschnittsgrösse, -gewicht und BMI aller  übergewichtigen Mitglieder

average_overweighted = df_member_list.loc[
    df_member_list["BMI"] > 25, ["Grösse", "Gewicht", "BMI"]
].mean()
average_overweighted
```
