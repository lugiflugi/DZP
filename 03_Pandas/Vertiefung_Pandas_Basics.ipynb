# 1. Warum pandas?

In Life-Science-Projekten entstehen häufig tabellarische Daten:
- Fermentations-/Bioreaktor-Messreihen (Zeit, Temperatur, pH, OD, Substrate)
- Qualitätskontrolle in der Lebensmitteltechnologie (Temperaturbereiche, Grenzwerte)
- Chemische Analytik (Konzentrationen, Nachweisgrenzen)
- Health-Daten (Messreihen, Laborwerte; anonymisiert)

**pandas** ist das Standardwerkzeug in Python, um solche Tabellen effizient zu:
- laden (CSV, Excel),
- untersuchen (head/info/describe),
- filtern/sortieren,
- und wieder zu exportieren.

Pandas ist eine Python-Bibliothek, die häufig für die Datenanalyse, -manipulation und -verarbeitung verwendet wird. Es bietet eine einfache und leistungsfähige Möglichkeit, strukturierte Daten zu verarbeiten und zu analysieren, insbesondere für in Tabellen organisierte Daten.

Pandas basiert auf den Konzepten von Series und DataFrame, die es ermöglichen, Daten in tabellarischer Form zu organisieren und zu manipulieren. Eine Series ist eine eindimensionale Datenstruktur, ähnlich wie ein Array oder eine Liste, die einen Index und eine Sequenz von Werten enthält. Ein DataFrame ist eine zweidimensionale Datenstruktur, die als Tabelle dargestellt werden kann. Es enthält mehrere Spalten, jede davon repräsentiert eine Series, und einen Index, der die Zeilen der Tabelle darstellt.

## 3.2 DataFrame
Ein DataFrame ist eine der wichtigsten Datenstrukturen in der Pandas-Bibliothek und kann als eine zweidimensionale Tabelle betrachtet werden, die aus Zeilen und Spalten besteht. Jede Spalte im DataFrame kann als eine Series betrachtet werden, die einen eindeutigen Namen und eine Sequenz von Werten enthält, während der Index die Zeilen des DataFrames repräsentiert.

Ein **DataFrame** ist zweidimensional: mehrere Spalten.

# 12 Zusammenführen von DataFrames mit .merge()
Die Methode `.merge()` ermöglicht es, zwei DataFrames basierend auf gemeinsamen Spalten oder Indizes zu kombinieren. Dies ist besonders nützlich, wenn Sie Daten aus verschiedenen Quellen zusammenführen möchten, die gemeinsame Schlüssel haben.

- pd.merge(df1, df2, on='gemeinsame_spalte', how='inner')  # Inner Join
  - how 'left': Alle Zeilen aus df1 + passende aus df2
  - how 'right': Alle Zeilen aus df2 + passende aus df1
  - how 'outer': Alle Zeilen aus beiden DataFrames, fehlende Werte mit NaN auffüllen
  - how 'inner': Nur Zeilen, die in beiden DataFrames übereinstimmen

Schauen wir ein Beispiel:
- Dataframe1 : Folgende Spalten: PatientenID, Gewicht, Grösse und Alter, Raucher (dummywerte aus der Funktion generate_patient_data_source1(n=100))
- Dataframe2 : Folgende Spalten: PatientenID,Geschlecht, Blutgruppe und Rhesusfaktor (dummywerte aus der Funktion generate_patient_data_source2(n=100))

Ziel: eine Dataframe mit allen Informationen zu den Patienten (PatientenID, Gewicht, Grösse, Alter, Raucher, Geschlecht, Blutgruppe und Rhesusfaktor) erstellen.

```python
# Funktionen zum Generien von Patientendaten
import random


def generate_patient_data_source1(num_patients):
    patients = []
    for i in range(num_patients):
        patient = {
            "PatientID": i,
            "Alter": random.randint(20, 80),
            "Grösse": random.randint(150, 200),
            "Gewicht": random.randint(50, 120),
            "Raucher": random.choice([True, False]),
        }
        patients.append(patient)
    return patients


def generate_patient_data_source2(num_patients):
    patients = []
    for i in range(num_patients):
        patient = {
            "PatientID": i,
            "Geschlecht": random.choice(["m", "w"]),
            "Blutgruppe": random.choice(["A", "B", "AB", "0"]),
            "Rhesusfaktor": random.choice(["+", "-"]),
        }
        patients.append(patient)
    return patients
```
