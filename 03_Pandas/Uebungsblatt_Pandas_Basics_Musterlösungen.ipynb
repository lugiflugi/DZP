d. Sie haben zwei neue Listen erhalten. In der einen Liste sind die Mitglieder, die eine Mitgliedschaft gekündigt haben und in der anderen Liste sind die Mitglieder, die eine Mitgliedschaft abgeschlossen haben. Lesen Sie die Dateien `ko_fitness_memberlist_canceled.csv` und `ko_fitness_memberlist_new.csv` ein und ergänzen Sie die Informationen in `ko_fitness_memberlist`. Die Mitglieder sind über die Spalten `member_id` eindeutig identifizierbar.

```python
# Liste mit gekündigten Mitgliedern einlesen
ko_fitness_memberlist_cancelled = pd.read_csv(
    "ko_fitness_memberlist_cancelled.csv", sep=";", encoding="utf-8"
)
ko_fitness_memberlist_cancelled
```

```python
# Mitglieder, die gekündigt haben, aus der Liste entfernen mit for-Schleife
for member_id in ko_fitness_memberlist_cancelled["member_id"]:
    ko_fitness_memberlist = ko_fitness_memberlist[
        ko_fitness_memberlist["member_id"] != member_id
    ]

# Mitglieder, die gekündigt haben, aus der Liste entfernen mit isin()
ko_fitness_memberlist = ko_fitness_memberlist[
    ~ko_fitness_memberlist["member_id"].isin(
        ko_fitness_memberlist_cancelled["member_id"]
    )
]

ko_fitness_memberlist
```

```python
# neue Mitglieder hinzufügen
ko_fitness_memberlist_new = pd.read_csv(
    "ko_fitness_memberlist_new.csv", sep=";", encoding="utf-8"
)
ko_fitness_memberlist_new.head()
```

```python
# Mitgliederliste um neue Mitglieder erweitern
ko_fitness_memberlist = pd.concat(
    [ko_fitness_memberlist, ko_fitness_memberlist_new], ignore_index=True
)
ko_fitness_memberlist
```

e. Und nun berechnen Sie die BMI und BMI-Kategorie für jedes Mitglied und speichern Sie die Ergebnisse in `ko_fitness_memberlist`. Der BMI wird berechnet als Gewicht in Kilogramm geteilt durch das Quadrat der Grösse in Metern. Die BMI-Kategorien sind:

```python
# Funktion bmi_message
def bmi_message(bmi):
    return_val = "Unter 18.5: Untergewicht."
    if bmi > 40:
        return_val = "Über 40: Adipositas Grad III."
    elif bmi >= 35:
        return_val = "35 - 39.9: Adipositas Grad II."
    elif bmi >= 30:
        return_val = "30 - 34.9: Adipositas (Fettleibigkeit) Grad I."
    elif bmi >= 25:
        return_val = "25 - 29.9: Übergewicht/ Präadipositas."
    elif bmi >= 18.5:
        return_val = "18.5 - 24.9: Normalgewicht."
    return return_val


# Funktion bmi_calc
def bmi_calc(weight_in_kg, height_in_m):
    return round(weight_in_kg / height_in_m**2, 2)


# BMI & BMI_Message berechnen mit for-Schleife
bmi_list = []
bmi_message_list = []
for index, row in ko_fitness_memberlist.iterrows():
    bmi = bmi_calc(row["Gewicht"], row["Grösse"])
    bmi_list.append(bmi)
    bmi_message_list.append(bmi_message(bmi))
ko_fitness_memberlist["BMI"] = bmi_list
ko_fitness_memberlist["BMI_Message"] = bmi_message_list

ko_fitness_memberlist

# Das ganze könnte man elegentar gestalten undzwar mit der apply() Methode.
# Dazu werden wir in den nächsten Wochen mehr erfahren

# ko_fitness_memberlist["BMI"] = ko_fitness_memberlist.apply(lambda row: bmi_calc(row["Gewicht"], row["Grösse"]), axis=1)
# ko_fitness_memberlist["BMI_Message"] = ko_fitness_memberlist["BMI"].apply(bmi_message)

# ko_fitness_memberlist
```

i. 

Was ist das durschnittliche Gewicht und die kleineste und grösste Körpergrösse?
- [.mean](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.mean.html)
- [.min](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.min.html)
- [.max](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.max.html)

Wie viele Mitglieder sind männlich und wie viele weiblich? 
- [.value_counts](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html)

Wie viele einzigartige Nachnamen gibt es? 
- [.nunique](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.nunique.html)

Und was ist der am häufigsten vorkommende Vorname?
- [.value_counts](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html)
- [.idxmax](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.idxmax.html)


```python
# - durchscnittliches Gewicht: 81.75 kg
mean_weight = ko_fitness_memberlist["Gewicht"].mean()
print(f"Das durchschnittliche Gewicht beträgt: {mean_weight} kg")

# - kleinste Körpergrösse: 1.51 m
min_height = ko_fitness_memberlist["Grösse"].min()
print(f"Die kleinste Körpergrösse beträgt: {min_height} m")

# - grösste Körpergrösse: 2.00 m
max_height = ko_fitness_memberlist["Grösse"].max()
print(f"Die grösste Körpergrösse beträgt: {max_height} m")

# - Anzahl männlich: 73
# - Anzahl weiblich: 60
print(ko_fitness_memberlist["Geschlecht"].value_counts())

# - Anzahl einzigartige Namen: 100
unique_names = ko_fitness_memberlist["Name"].nunique()
print(f"Anzahl einzigartige Namen: {unique_names}")

# - am häufigsten vorkommender Vorname: Hans
most_common_name = ko_fitness_memberlist["Vorname"].value_counts().idxmax()
print(f"am häufigsten vorkommender Vorname: {most_common_name}")
```
