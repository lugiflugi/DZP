## 1) Problemorientierter Einstieg

Messwerte einer Produktionslinie:

````python
messwerte = [21.3, 19.8, 23.1, 18.4, 25.0, 22.6]
messwerte
````

### Aufgabe
Filtere alle Werte > 22.

### Variante A – klassische Schleife

````python
ueber_22 = []
for m in messwerte:
    if m > 22:
        ueber_22.append(m)
ueber_22
````

### Variante B – List Comprehension

````python
ueber_22 = [m for m in messwerte if m > 22]
ueber_22
````

**Reflexion:**
- Was ist strukturell anders? --> kompakter, weniger Zeilen, direkter Ausdruck der Logik
- Welche Variante ist kompakter? --> List Comprehension

## 2) Einführung in List Comprehensions
List Comprehensions sind eine elegante und effiziente Möglichkeit, um Listen in Python zu erstellen. Sie ermöglichen es, eine neue Liste zu generieren, indem man eine bestehende Liste durchläuft und dabei Elemente transformiert oder filtert. Die allgemeine Syntax lautet:

```python
[expression for item in iterable if condition]
```

- **expression**: Der Ausdruck, der auf jedes Element angewendet wird, um das neue Element zu erzeugen. --> Transformation
- **item**: Das aktuelle Element aus der iterierbaren Sequenz. --> Iteration
- **iterable**: Eine Sequenz (z. B. Liste, Tuple), die durchlaufen wird. --> Iteration
- **condition** (optional): Eine Bedingung, die bestimmt, ob das Element in die neue Liste aufgenommen wird. --> optionale Selektion


## 3) Praxisnahe Übungssequenz – Produktionskontext

Gegeben sind tägliche Produktionsmengen (in kg) einer Anlage:

````python
produktion_kg = [480, 510, 495, 530, 470, 515, 505]
produktion_kg
````

### Aufgabe 1 – Transformation
Berechne für jeden Tag die Abweichung vom Zielwert (500 kg).

````python
abweichung = [p - 500 for p in produktion_kg]
abweichung
````

### Aufgabe 2 – Filterung
Filtere alle Tage mit Unterproduktion (< 490 kg).

````python
unterproduktion = [p for p in produktion_kg if p < 490]
unterproduktion
````

### Aufgabe 3 – Kombination
Erzeuge Labels ('ok' / 'kritisch'), wenn Produktion < 490 kg.

````python
labels = ['kritisch' if p < 490 else 'ok' for p in produktion_kg]
labels
````

## 4) Praxisbeispiel: Qualitäts-/Prozessdaten filtern

Beispiel: Messreihe eines Qualitätsindikators (z. B. Wasseraktivität `aw`) pro Charge. Wir markieren Chargen unterhalb eines Schwellenwerts als kritisch.

````python
aw_values = [0.96, 0.94, 0.89, 0.91, 0.87, 0.93, 0.95, 0.88, 0.90, 0.97]
kritisch_schwelle = 0.90
aw_values, kritisch_schwelle
````

### Variante A – for-Schleife

````python
kritisch = []
for v in aw_values:
    if v < kritisch_schwelle:
        kritisch.append(v)
kritisch
````

### Variante B – List Comprehension

````python
kritisch = [v for v in aw_values if v < kritisch_schwelle]
kritisch
````

## 5) Strukturierte Daten (Dictionary/JSON-ähnlich)

Beispiel-Datensatz: Temperatur pro Charge.

````python
daten = [
    {"charge": 1, "temp": 21.3},
    {"charge": 2, "temp": 23.4},
    {"charge": 3, "temp": 19.7}
]
daten
````

### Aufgaben
1. Extrahiere alle Temperaturen.
2. Filtere alle Chargen mit `temp > 22`. 
3. Erzeuge eine Liste mit Abweichung vom Mittelwert.

````python
temps = [d["temp"] for d in daten]
temps
````

````python
warme_chargen = [d["charge"] for d in daten if d["temp"] > 22]
warme_chargen
````

````python
mittel = sum(temps)/len(temps)
abweichung = [round(t - mittel, 2) for t in temps]
mittel, abweichung
````

## 6) Liste von Tupeln als Rückgabe + `zip()`

Oft liegen zusammengehörige Informationen in **zwei Listen** vor (z. B. Name der Charge und Messwert). Wir können daraus eine Liste von Tupeln erzeugen.

````python
chargen = ["A1", "A2", "B1"]
werte = [91.2, 87.5, 93.1]
chargen, werte
````

### Variante A – Index-basiert

````python
paare = [(chargen[i], werte[i]) for i in range(len(chargen))]
paare
````

### Variante B – eleganter mit `zip()`
`zip()` verknüpft Elemente gleicher Position zu Tupeln.

````python
paare_zip = list(zip(chargen, werte))
paare_zip
````

## 7) Verschachtelte Comprehensions – Mehrdimensionale Prozessdaten

Gegeben sind Temperaturmessungen pro Tag (3 Schichten pro Tag):

````python
temperaturen = [
    [21.3, 22.1, 20.9],  # Tag 1
    [23.4, 24.0, 22.8],  # Tag 2
    [19.7, 20.2, 18.9]   # Tag 3
]
temperaturen
````

### Aufgabe 1 – Alle Messwerte in eine Liste überführen (Flatten)

````python
alle_werte = [t for tag in temperaturen for t in tag]
alle_werte
````

### Aufgabe 2 – Alle Messwerte > 22°C extrahieren

````python
kritische_werte = [t for tag in temperaturen for t in tag if t > 22]
kritische_werte
````

### Analyseauftrag
- Welche Schleife wird zuerst ausgeführt?
- Warum steht die äussere Schleife zuerst?
- Wann wäre hier eine normale Schleife lesbarer?

### Musterantworten – Analyseauftrag zu verschachtelten Comprehensions

**Welche Schleife wird zuerst ausgeführt?**
- zuerst die **äussere** Iteration (z. B. `for tag in temperaturen`), dann pro äusserem Element die **innere** (`for t in tag`).

**Warum steht die äussere Schleife zuerst?**
- Syntax folgt der Lesereihenfolge: *"für jedes Tag-Subarray … für jeden Messwert …"*

**Wann wäre eine normale Schleife lesbarer?**
- wenn zusätzliche Bedingungen/Fehlerbehandlung (Missing Values, Ausreisserlogik) nötig sind
- wenn mehrere Zwischenschritte erklärt/geloggt werden sollen
- wenn Studierende die Logik noch nicht sicher mental „entfalten“ können (didaktische Entscheidung).

## 8) Performance-Vergleich

List Comprehensions sind oft schneller, aber **nicht immer** der wichtigste Faktor. Wir messen hier exemplarisch.

Hinweis: `%%timeit` ist ein *Jupyter magic command* und misst die Ausführungszeit mehrmals.

### Musterantworten – Performance-Block

- List Comprehensions sind oft schneller, weil sie in Python intern effizienter umgesetzt werden.
- In realen Projekten ist der Unterschied häufig kleiner als der Einfluss von **I/O**, **pandas/NumPy**, oder Datenmenge.
- Entscheidungskriterium ist meist: **Lesbarkeit/Wartbarkeit zuerst**, Performance nur bei Bedarf messen.

````python
import random
zahlen = [random.randint(1, 100) for _ in range(10000)]
def loop_version(xs):
    out = []
    for x in xs:
        if x % 2 == 0:
            out.append(x * x)
    return out

def comp_version(xs):
    return [x*x for x in xs if x % 2 == 0]

len(zahlen)
````

### Zeitmessung – List Comprehension

````python
%%timeit -r 4 -n 10
comp_version(zahlen)
````

### Zeitmessung – for-Schleife

````python
%%timeit -r 4 -n 10
loop_version(zahlen)
````

**Interpretation:**
- Unterschiede sind oft klein, aber messbar.
- In der Praxis sind **Lesbarkeit** und **Wartbarkeit** häufig wichtiger als Mikro-Optimierungen.

## 10) Vor- und Nachteile von Comprehensions

**Vorteile:**
- kompakter Code
- oft schneller als eine explizite for-Schleife
- gut lesbar bei einfachen Operationen

**Nachteile:**
- bei vielen Verschachtelungen sinkt Lesbarkeit
- Debugging ist schwieriger (Zwischenschritte fehlen)

**Merksatz:** *So kurz wie möglich – so klar wie nötig.*

## Abschlussreflexion

- Wann ist eine List Comprehension **nicht** sinnvoll?
- Ab welcher Komplexität leidet Lesbarkeit?
- Wo in einer Datenpipeline ist eine Comprehension besonders nützlich (vor `pandas`/DataFrame)?

### Musterantworten – Abschlussreflexion

**1) Wann ist eine List Comprehension nicht sinnvoll?**
- wenn die Logik mehrstufig/komplex ist (mehrere Bedingungen, viele Sonderfälle)
- wenn Zwischenschritte fürs Debugging/Logging nötig sind
- wenn Seiteneffekte auftreten (z. B. Datei-/DB-Operationen)
- wenn die Lesbarkeit deutlich leidet (extreme Verschachtelung und/oder ternäre Ausdrücke)

**2) Ab welcher Komplexität leidet Lesbarkeit?**
- häufig ab **mehr als einer Verschachtelung** oder bei langen Bedingungen mit `and`/`or`
- wenn man die Comprehension nicht mehr „in einem Satz“ erklären kann
- wenn Transformation + Filter + ternärer Ausdruck gleichzeitig vorkommen (terneren Ausdruck bedeutet: `x if condition else y`)

**3) Wo in einer Datenpipeline ist sie besonders nützlich?**
- beim **Vorstrukturieren** von Rohdaten (Listen/Dicts), z. B. aus CSV/JSON/API
- beim schnellen Extrahieren/Filtern einzelner Felder *vor* `pandas`
- beim Erzeugen einfacher Features/Labels (z. B. "gut"/"kritisch")

**Merksatz:** *So kurz wie möglich – so klar wie nötig.*
