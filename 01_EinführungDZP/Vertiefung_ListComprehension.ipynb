## 2) Einführung in List Comprehensions
List Comprehensions sind eine elegante und effiziente Möglichkeit, um Listen in Python zu erstellen. Sie ermöglichen es, eine neue Liste zu generieren, indem man eine bestehende Liste durchläuft und dabei Elemente transformiert oder filtert. Die allgemeine Syntax lautet:

```python
[expression for item in iterable if condition]
```

- **expression**: Der Ausdruck, der auf jedes Element angewendet wird, um das neue Element zu erzeugen. --> Transformation
- **item**: Das aktuelle Element aus der iterierbaren Sequenz. --> Iteration
- **iterable**: Eine Sequenz (z. B. Liste, Tuple), die durchlaufen wird. --> Iteration
- **condition** (optional): Eine Bedingung, die bestimmt, ob das Element in die neue Liste aufgenommen wird. --> optionale Selektion

## 8) Performance-Vergleich

List Comprehensions sind oft schneller, aber **nicht immer** der wichtigste Faktor. Wir messen hier exemplarisch.

Hinweis: `%%timeit` ist ein *Jupyter magic command* und misst die Ausführungszeit mehrmals.

````python
import random
zahlen = [random.randint(1, 100) for _ in range(10000)]
def loop_version(xs):
    out = []
    for x in xs:
        if x % 2 == 0:
            out.append(x * x)
    return out

def comp_version(xs):
    return [x*x for x in xs if x % 2 == 0]

len(zahlen)
````

### Zeitmessung – List Comprehension

````python
%%timeit -r 4 -n 10
comp_version(zahlen)
````

### Zeitmessung – for-Schleife

````python
%%timeit -r 4 -n 10
loop_version(zahlen)
````

**Interpretation:**
- Unterschiede sind oft klein, aber messbar.
- In der Praxis sind **Lesbarkeit** und **Wartbarkeit** häufig wichtiger als Mikro-Optimierungen.

## 10) Vor- und Nachteile von Comprehensions

**Vorteile:**
- kompakter Code
- oft schneller als eine explizite for-Schleife
- gut lesbar bei einfachen Operationen

**Nachteile:**
- bei vielen Verschachtelungen sinkt Lesbarkeit
- Debugging ist schwieriger (Zwischenschritte fehlen)

**Merksatz:** *So kurz wie möglich – so klar wie nötig.*
