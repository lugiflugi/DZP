## Übung 5 – Dictionary-Liste → Tupel + Filter + Sortierung

Ein Weinkeller verwaltet Weine als Liste von Dictionaries:

```python
weine = [
    {'region': 'Wallis', 'weingut': 'Muster',  'jahrgang': 2012},
    {'region': 'Waadt',  'weingut': 'Example', 'jahrgang': 2018},
    {'region': 'Tessin', 'weingut': 'Casa',    'jahrgang': 2015},
    {'region': 'Wallis', 'weingut': 'Alpin',   'jahrgang': 2019},
    {'region': 'Waadt',  'weingut': 'CaveX',   'jahrgang': 2010},
]
```

**Aufgaben:**

a) Erstelle eine Liste von Tupeln `(region, weingut, jahrgang)` für alle Weine mit `jahrgang > 2014`.

b) Erstelle dieselbe Tupel-Liste, **sortiert nach Jahrgang absteigend**.

c) Reflexion: Wann würdest du statt einer Comprehension lieber eine for-Schleife verwenden (mind. 2 Kriterien)?

Hinweis für b): `sorted(..., key=..., reverse=True)`

````python
weine = [
    {"region": "Wallis", "weingut": "Muster", "jahrgang": 2012},
    {"region": "Waadt", "weingut": "Example", "jahrgang": 2018},
    {"region": "Tessin", "weingut": "Casa", "jahrgang": 2015},
    {"region": "Wallis", "weingut": "Alpin", "jahrgang": 2019},
    {"region": "Waadt", "weingut": "CaveX", "jahrgang": 2010},
]
weine
````

````python
tupel_liste = [
    (w["region"], w["weingut"], w["jahrgang"]) for w in weine if w["jahrgang"] > 2014
]
tupel_liste
````

````python
tupel_sortiert = sorted(tupel_liste, key=lambda t: t[2], reverse=True)
tupel_sortiert
````

### Musterantwort (Reflexion)

- for-Schleife besser bei komplexer Logik/mehreren Sonderfällen
- besser, wenn Debugging/Logging/Zwischenschritte nötig
- besser, wenn Lesbarkeit bei langen Bedingungen/Verschachtelung leidet
- Wartbarkeit/Erweiterbarkeit als Argument

**Merksatz:** *So kurz wie möglich – so klar wie nötig.*

## Übung 6 – Patientendaten + BMI + Performance

**Kontext:** synthetische Datenübung (keine medizinische Beratung).

Wir simulieren viele Datensätze. Passe `N` an deinen Rechner an (z. B. 200_000 bis 2_000_000).

````python
import random

N = 300_000
patienten = [
    {
        "alter": random.randint(18, 90),
        "groesse_cm": random.randint(150, 200),
        "gewicht_kg": random.randint(45, 120),
        "raucher": random.choice([True, False]),
    }
    for _ in range(N)
]
len(patienten)
````

**Kriterium Hochrisiko:** `alter > 50` UND `BMI > 25` UND `raucher == True`.

**Aufgaben:**

a) Erzeuge eine Liste `bmis` mit dem BMI pro Person.

b) Erzeuge `hochrisiko_loop` mit einer for-Schleife.

c) Erzeuge `hochrisiko_comp` mit einer List Comprehension.

d) Vergleiche Laufzeiten mit `%%timeit` (am besten über Funktionen).

e) Beantworte Analysefragen: schneller? warum? Einfluss von `N`? Lesbarkeit?

````python
bmis = [p["gewicht_kg"] / ((p["groesse_cm"] / 100) ** 2) for p in patienten]
bmis[:5]
````

````python
hochrisiko_loop = []
for p in patienten:
    bmi = p["gewicht_kg"] / ((p["groesse_cm"] / 100) ** 2)
    if p["alter"] > 50 and bmi > 25 and p["raucher"]:
        hochrisiko_loop.append(p)
len(hochrisiko_loop)
````

````python
hochrisiko_comp = [
    p
    for p in patienten
    if p["alter"] > 50
    and (p["gewicht_kg"] / ((p["groesse_cm"] / 100) ** 2)) > 25
    and p["raucher"]
]
len(hochrisiko_comp)
````

### Performance (Jupyter)

Für faire Messung kapseln wir die Varianten in Funktionen und messen mit `%%timeit`.

````python
def make_loop(xs):
    out = []
    for p in xs:
        bmi = p["gewicht_kg"] / ((p["groesse_cm"] / 100) ** 2)
        if p["alter"] > 50 and bmi > 25 and p["raucher"]:
            out.append(p)
    return out


def make_comp(xs):
    return [
        p
        for p in xs
        if p["alter"] > 50
        and (p["gewicht_kg"] / ((p["groesse_cm"] / 100) ** 2)) > 25
        and p["raucher"]
    ]


# In Jupyter ausführen:
# %%timeit -r 3 -n 3
# make_loop(patienten)

# %%timeit -r 3 -n 3
# make_comp(patienten)
````

````python
%%timeit -r 3 -n 3
make_comp(patienten)
````

````python
%%timeit -r 3 -n 3
make_loop(patienten)
````

### Musterantworten (Analyse)

- **Schneller?** Oft ist die Comprehension etwas schneller (weniger Python-Overhead), aber nicht immer dramatisch.
- **Einfluss von N:** Laufzeit wächst ungefähr linear; Unterschiede werden bei grossem `N` messbarer.
- **Lesbarkeit:** Bei einfachen Bedingungen ok; bei komplexen Bedingungen/Feature-Berechnung ist die Schleife oft klarer (Zwischenschritte).

## Übung 7 – Verschachtelte Comprehensions (Mehrdimensionale Daten)

Sensorwerte (3 Schichten pro Tag):

```python
sensorwerte = [
    [21.3, 22.1, 20.9],
    [23.4, 24.0, 22.8],
    [19.7, 20.2, 18.9]
]
```

**Aufgaben:**

a) Erzeuge eine flache Liste aller Werte.

b) Extrahiere alle Werte `> 22`.

c) Berechne den Durchschnitt aller Werte (Generator Expression erlaubt).

````python
sensorwerte = [
    [21.3, 22.1, 20.9],
    [23.4, 24.0, 22.8],
    [19.7, 20.2, 18.9]
]
sensorwerte
````

````python
alle = [t for tag in sensorwerte for t in tag]
alle
````

````python
kritisch = [t for tag in sensorwerte for t in tag if t > 22]
kritisch
````

````python
avg = sum(t for tag in sensorwerte for t in tag) / sum(1 for tag in sensorwerte for _ in tag)
avg
````

## Übung 8 – Anspruchsvoll: Mehrere Bedingungen + Bewertung

Daten (Charge, Temperatur, pH):

```python
daten = [
    {'charge': 1, 'temp': 23.1, 'ph': 4.5},
    {'charge': 2, 'temp': 19.8, 'ph': 4.8},
    {'charge': 3, 'temp': 24.2, 'ph': 4.3},
    {'charge': 4, 'temp': 21.0, 'ph': 4.6},
]
```

**Aufgaben:**

a) Erzeuge eine Liste der `charge`-Nummern, bei denen `temp > 22` **und** `ph < 4.6`.

b) Analyse: Ist die Comprehension gut lesbar? Begründe (mind. 2 Punkte).

````python
daten = [
    {'charge': 1, 'temp': 23.1, 'ph': 4.5},
    {'charge': 2, 'temp': 19.8, 'ph': 4.8},
    {'charge': 3, 'temp': 24.2, 'ph': 4.3},
    {'charge': 4, 'temp': 21.0, 'ph': 4.6},
]
daten
````

````python
kritische_chargen = [d['charge'] for d in daten if d['temp'] > 22 and d['ph'] < 4.6]
kritische_chargen
````

### Musterantwort (Lesbarkeit)

- Gut lesbar, weil nur **eine** Bedingungskombination und klare Feldnamen.
- Wird schnell unlesbar bei vielen Bedingungen, verschachtelten Comprehensions oder wenn komplexe Berechnungen im Ausdruck stecken.
- Alternative: Schleife mit benannten Zwischenvariablen (Debugging/Wartbarkeit).
