# Peer-Feedback zur Präsentation und Projektarbeit

Das Peer-Feedback ist ein wichtiger Bestandteil der Lern- und Weiterentwicklungsprozesse im Rahmen der Projektarbeit. Nutzt die folgenden Fragen und Hinweise, um euer Feedback strukturiert und konstruktiv zu gestalten.

---

## Bewertungsfragen

Beurteilt folgende Aspekte in eurem Feedback:

- **Verständlichkeit:** Ist das Projekt verständlich und klar dargestellt?
- **Klarheit der Ergebnisse:** Sind die präsentierten Ergebnisse klar und nachvollziehbar dargestellt?
- **Korrektheit der Ergebnisse:** Sind die präsentierten Ergebnisse plausibel und fachlich korrekt?
- **Methoden und Modelle:** Wurden die Methoden und Modelle passend und korrekt angewendet?
- **Datenquellen:** Sind die verwendeten Datenquellen passend und korrekt gewählt?

---

## Regeln für konstruktives Peer-Feedback

Beim Erstellen eurer Rückmeldungen achtet bitte auf folgende Grundsätze:

- **Spezifisch und präzise:**  
  Vermeidet Verallgemeinerungen und gebt konkrete Beispiele an.

- **Konstruktiv und konkret:**  
  Bietet konkrete Handlungsmöglichkeiten und Verbesserungsvorschläge an.

- **Optimistisch und unterstützend:**  
  Gebt Feedback, das die Weiterentwicklung fördert (Coaching-Ansatz).

- **Respektvoll:**  
  Achtet auf einen wertschätzenden Umgang und vermeidet Blossstellungen.

- **Zeitgerecht:**  
  Gebt Feedback zeitnah.

- **Sorgfältig:**  
  Überprüft eure Kommentare sorgfältig auf Verständlichkeit und Fairness.

- **Positiv und ausgewogen:**  
  Benennt sowohl gelungene Aspekte als auch verbesserungswürdige Punkte.  
  Konzentriert euch auf die Leistung, nicht auf persönliche Merkmale.

## Feedback nach der 3-2-1-Methode

Nutzt diese Methode, um strukturiert Feedback zu geben:

- **3 Dinge**, die ihr aus der Präsentation gelernt habt.
- **2 konkrete Vorschläge**, wie die Präsentation inhaltlich verbessert werden könnte.
- **1 offene Frage**, die ihr nach der Präsentation noch habt.

## Hinweise für Feedback-Empfänger:innen

Beim Erhalten von Feedback gilt:

- Konzentriert und aufmerksam zuhören bzw. lesen.
- Macht euch Notizen.
- Rechtfertigungen oder Erklärungen vermeiden.
- Feedback nicht persönlich nehmen; auf emotionale Reaktionen achten.
- Entscheidet selbst, welche Anregungen ihr umsetzen möchtet.
