# Leistungsnachweis: DZP-Projekt  
## Datenverarbeitungspipeline im Life-Sciences-Bereich


Im Rahmen der Gruppenarbeit soll eine **Datenverarbeitungspipeline** entwickelt werden, die Daten aus einer **Web-API** bezieht, transformiert, analysiert und visualisiert. Die Ergebnisse sollen in einer geeigneten Form gespeichert werden. Der Schwerpunkt liegt auf der Erstellung einer **Datenverarbeitungspipeline in Python**. Dabei sollen insbesondere die Packages **Pandas** für die Datenverarbeitung und **Seaborn** für die Visualisierung eingesetzt werden.

Die Gruppenarbeit orientiert sich an den im Modul *Datenzentriertes Programmieren* geforderten Kompetenzen:

- Nutzung kombinierter Datentypen und Organisation der Daten in DataFrames
- Anwendung funktionaler Programmieransätze, sowie den Grundprinzipien der Datenzentrierten Programmierung (Folien SW1)
- Explorative Datenanalyse und Visualisierung strukturierter Daten
- Reflexion über die Arbeitsweise und Umsetzung in kleinen Teams

---

## Aufgabenstellung

**Datenquelle und Fragestellung**
   - Wählt eine passende Web-API aus dem Bereich **Life Sciences** (z.B. Umwelt, Ernährung, Gesundheit). Hier eine Liste von möglichen APIs [`Beispiele_APIs.ipynb`](https://github.zhaw.ch/ADLS-Data-Science-and-Computation/Datenzentriertes-Programmieren-FS26/blob/main/00_Gruppenarbeit/Beispiele_APIs.ipynb).
   - Alternativ könnt ihr eine eigene API finden, die für eure Fragestellung geeignet ist.
   - Formuliert eine präzise **Fragestellung**, die mit den ausgewählten Daten beantwortet werden kann.
   - Als Orientierung dienen die Notebooks [`Beispiel_Projekt_arxiv.ipynb`](https://github.zhaw.ch/ADLS-Data-Science-and-Computation/Datenzentriertes-Programmieren-FS26/blob/main/00_BeispielProjekte/Beispiel_Projekt_arxiv.ipynb) und [`Beispiel_Projekt_Wetteranalyse.ipynb`](https://github.zhaw.ch/ADLS-Data-Science-and-Computation/Datenzentriertes-Programmieren-FS26/blob/main/00_BeispielProjekte/Beispiel_Projekt_Wetteranalyse.ipynb) aus der Einführungswoche.
   - Stimmt die Fragestellung und Datenquelle vor Beginn mit dem Dozenten ab.

**Datenpipeline erstellen**
   - Bindet die API erfolgreich an (oder ladet die CSV-Datei).
   - Transformiert die Rohdaten in **generische Datentypen** (falls nötig).
   - Wählt die relevanten Daten aus und speichert die Daten in einem **Pandas DataFrame**.
   - Führt eine **explorative Analyse** der Daten durch.
   - Visualisiert die Ergebnisse mit **Seaborn**.
   - Speichert die Resultate und Visualisierungen in geeigneter Form.

**Reflexion und Dokumentation**
   - Dokumentiert die einzelnen Schritte der Datenverarbeitung ausführlich.
   - Reflektiert die Teamarbeit, Rollenverteilung und individuelle Beiträge (am Ende des Notebooks).

**Präsentation (10–12 Minuten)**
   - Einleitung und Motivation
   - Beschreibung des Projekts und der verwendeten Methoden
   - Vorstellung der Datenquellen
   - Präsentation der Ergebnisse und Diskussion
   - Fazit und Ausblick

**Peer-Feedbaack**
   - Gebt euch gegenseitig Feedback zu den Präsentationen und der Teamarbeit.
   - Reflektiert die Rückmeldungen und besprecht diese im Team.

## Anforderungen an die Projektarbeit

### 1. Datenpipeline
#### 1.1 Funktionalität (30 Punkte)

| Kriterium                      | Punkte |
|--------------------------------|--------|
| Datenquelle                | 5      |
| Datentransformation        | 10     |
| Explorative Analyse        | 5      |
| Visualisierungen (Seaborn) | 5      |
| Datenexport                | 5      |

**Details zur Funktionalität:**

- **Datenquelle**  
  - Erfolgreiche Anbindung einer Web-API oder eines PyPI-Pakets  
  - Alternativ CSV-Datei möglich, dann jedoch mit Punktabzug
- **Datentransformation**  
  - Transformation der Rohdaten in generische Datentypen  
  - Auswahl relevanter Daten  
  - Speicherung der Daten in einem Pandas DataFrame  
  - Anreicherung mit zusätzlichen Informationen (optional)
- **Explorative Analyse**  
  - Analyse der Daten mittels Pandas
- **Visualisierung**  
  - Aussagekräftige statistische Visualisierungen mit Seaborn
- **Datenexport**  
  - Speicherung der Ergebnisse (Daten und Visualisierungen) in geeigneter Form

---
