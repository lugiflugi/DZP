### This file needs to be run only once to create the CSV files for each country

```python
import geonamescache
import csv

gc = geonamescache.GeonamesCache()

# Fetch all countries from geonamescache
countries = gc.get_countries()

# Fetch City data with the ISO codes of all countries in Europe (including Turkey, Russia and Kazakhstan)
european_country_codes = ['AL', 'AD', 'AM', 'AT', 'AZ', 'BY', 'BE', 'BA', 'BG', 'HR', 'CY', 'CZ', 'DK', 'EE', 'FO', 'FI', 'FR', 'GE', 'DE', 'GI', 'GR', 'GL', 'GG', 'HU', 'IS', 'IE', 'IM', 'IT', 'JE', 'KZ', 'XK', 'LV', 'LT', 'LU', 'MK', 'MT', 'MD', 'MC', 'ME', 'NL', 'NO', 'PL', 'PT', 'RO', 'RU', 'RS', 'SK', 'SI', 'ES', 'SE', 'CH', 'UA', 'GB', 'TR', 'UZ']
# Fetch all cities from geonamescache
cities = gc.get_cities()
```

```python
# Get European cities with population greater than 10,000
european_cities = []
for city_id, city_data in cities.items():
    # fetch all cities with greater than 10,000 population
    if city_data['countrycode'] in european_country_codes and city_data['population'] > 10000:
        # fetch city name, country code, latitude, longitude, population and geonameid
        city_name = city_data['name']
        city_country = city_data['countrycode']
        city_lat = city_data['latitude']
        city_lon = city_data['longitude']
        city_pop = city_data['population']
        city_id = city_data['geonameid']
        # append the city data to the list of european cities
        european_cities.append({'name': city_name, 'latitude': city_lat, 'longitude': city_lon, 'country': city_country, 'population': city_pop, 'id': city_id})
```

```python
# Save the European cities in separate CSV files for each country
for country_code in european_country_codes:
    country_cities = [city for city in european_cities if city['country'] == country_code]
    if country_cities:
        file_name = f"{country_code}_Cities.csv"
        with open(file_name, 'w', newline='', encoding='utf-8') as csvfile:
            # set column names
            fieldnames = ['name', 'latitude', 'longitude', 'country', 'population', 'id']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(country_cities)
            # print the file name for each country that has been created
            print(f"CSV file for {country_code} has been created.")
```
