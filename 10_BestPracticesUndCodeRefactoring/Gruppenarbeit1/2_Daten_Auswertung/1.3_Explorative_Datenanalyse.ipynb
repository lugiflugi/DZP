```python
# define function that filters sorts and cleans data frame

def df_read_clean(file_name):
    df = pd.read_csv(file_name)
    df = df.sort_values(by=['city'])
    df = df.dropna()
    df = df.drop_duplicates()
    df = df.drop(['id'], axis=1)
    df = round(df,2)
    return df
```

```python
# loop through every year with a for loop and create a data frame for every year
years = ['22', '21', '20', '19', '03', '90', '80', '70']
# select function
for year in years:
    filename = f"TEMP/TEMP_{year}.csv"
    globals()[f"df_{year}"] = df_read_clean(filename)

```

### Yearly average temperature across cities in Europe

```python
years = ['22', '21', '20', '19', '03', '90', '80', '70']

for year in years:
    filename = f"TEMP/TEMP_{year}.csv"
    globals()[f"df_temp_{year}"] = pd.read_csv(filename)

temp_list = ["df_temp_" + year for year in years]

# create a new column for yearly average in every city
for temp in temp_list:
    globals()[temp]['avg_temp'] = globals()[temp][['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug',
                                                   'sep', 'oct', 'nov', 'dec']].mean(axis=1)

# print the yearly average temperature for every year and print the whole year like 2022
for temp in temp_list:
    year = temp.split('_')[-1]  # Extract the year from the data frame name
    if year in ['70', '80', '90']:
        print(f"Year 19{year} average temperature: {round(globals()[temp]['avg_temp'].mean(), 2)} degrees celsius")
    else:
        print(f"Year 20{year} average temperature: {round(globals()[temp]['avg_temp'].mean(), 2)} degrees celsius")
```
