## 2) Einführung in `map()`, `filter()` & `lambda`

`map()` und `filter()` sind funktionale Werkzeuge, die Transformation und Auswahl klar trennen.

### Map()
Map() eine eingebaute Funktion in Python, die eine andere Funktion auf jedes Element eines iterierbaren Objekts wie einer Liste oder eines Tupels anwendet und einen Iterator zurückgibt, der die Ergebnisse enthält. Die map() Funktion ist nützlich, wenn Sie eine Funktion auf alle Elemente einer Liste anwenden möchten. Die gemappte Funktion wird nicht sofort ausgeführt, sondern erst, wenn sie benötigt wird, was bedeutet, dass sie nicht sofort eine Liste zurückgibt, sondern einen Iterator, der die Elemente enthält.

Link zur Dokumentation: (https://www.programiz.com/python-programming/methods/built-in/map)

### Filter()
Die Funktion filter() in Python ermöglicht es, eine Funktion auf jedes Element einer Sequenz (z. B. einer Liste oder eines Tupels) anzuwenden und nur diejenigen Elemente beizubehalten, für die die Funktion True zurückgibt. Das Ergebnis ist eine neue Sequenz, die nur aus den Elementen besteht, für die die Funktion True zurückgegeben hat. Die gefilterte Funktion wird nicht sofort ausgeführt, sondern erst, wenn sie benötigt wird, was bedeutet, dass sie nicht sofort eine Liste zurückgibt, sondern einen Iterator, der die Elemente enthält.

Link zu Dokumentation: (https://www.programiz.com/python-programming/methods/built-in/filter)

### Lambda-Funktionen
Eine Lambda-Funktion ist eine kleine anonyme Funktion. Eine Lambda-Funktion kann eine beliebige Anzahl von Argumenten haben, aber nur eine Ausdruck. Der Ausdruck wird ausgewertet und der Wert wird zurückgegeben. Lambda-Funktionen können in Kombination mit den Funktionen map() und filter() verwendet werden.

Link zu Dokumentation: (https://www.dataquest.io/blog/tutorial-lambda-functions-in-python/) 

Allgemeine Syntax:

```python
map(function, iterable)
filter(predicate, iterable)
```

- **function**: wird auf jedes Element angewendet (Transformation)
- **predicate**: liefert `True/False` (Filterung)
- beide liefern zunächst **Iteratoren** → häufig: `list(...)`


````python
numbers = [1, 2, 3, 4, 5]

a = list(map(lambda x: x * 2, numbers))
b = list(filter(lambda x: x > 3, numbers))
c = list(map(lambda x: x * 10, filter(lambda x: x > 3, numbers)))
a, b, c
````

## 5) Performance-Vergleich

Wir messen grob: for-loop vs list comp vs `filter()`+`map()`.

````python
data = list(range(100_000))
````

````python
%%timeit
r = []
for x in data:
    if x > 50_000:
        y = x * 2
        r.append(y)
````

````python
%%timeit
[x*2 for x in data if x > 50_000]
````

````python
%%timeit
list(map(lambda x: x*2, filter(lambda x: x > 50_000, data)))
````

## 8) List Comprehension vs map() & filter() Funktionen

Vorteile von List Comprehension:
- Lesbarkeit: Für einfache Ausdrücke sind List Comprehensions oft leichter zu lesen und zu verstehen als die entsprechenden map() oder filter() Aufrufe.
- Flexibilität: Erlauben die gleichzeitige Anwendung von Filter- und Transformationslogik in einer einzigen, klaren Zeile Code.
- Performance: In einigen Fällen können List Comprehensions schneller sein als die Kombination von map() und filter() wegen der spezifischen Python-Implementierung.

Vorteile von map() und filter():
- Performance: In einigen Fällen können map() und filter() schneller sein als List Comprehensions, insbesondere bei komplexen Funktionen.
- Wiederverwendbarkeit: map() und filter() können auf dieselbe Weise wie andere Funktionen wiederverwendet werden, was die Wartung und Lesbarkeit des Codes verbessert.
- Spezifität: Klar definiert für Filtrationsaufgaben bzw. Transformationsaufgaben. Seine Nutzung signalisiert deutlich die Absicht, Elemente zu filtern bzw. zu transformieren, was die Lesbarkeit des Codes verbessert.
- Laziness: Da filter() & map() in Python 3 einen Iterator zurückgibt, werden Elemente auf die "faule" Art verarbeitet (lazily evaluated), was bei der Verarbeitung grosser Datenmengen effizient sein kann.


## 9) Lambda-Funktionen vs def-Funktionen

Vorteile von Lambda-Funktionen:
- Kürze: Lambda-Funktionen sind kürzer und erfordern weniger Tippaufwand.
- Lesbarkeit: Lambda-Funktionen können in einer einzigen Zeile geschrieben werden, was die Lesbarkeit des Codes verbessert.
- Argument für Funktionen höherer Ordnung: Lambda-Funktionen können als Argumente für Funktionen höherer Ordnung wie map(), filter() und reduce() verwendet werden.
- Anonymität: Lambda-Funktionen sind anonym, was bedeutet, dass sie nicht mit einem Namen versehen werden müssen und daher in einer Zeile geschrieben werden können.

Vorteile von def-Funktionen:
- Mehrzeilige Funktionen: def-Funktionen können mehrere Zeilen Code enthalten, was die Lesbarkeit und Wartbarkeit des Codes verbessert.
- Wiederverwendbarkeit: def-Funktionen können wiederverwendet werden, was die Wartung und Lesbarkeit des Codes verbessert.
- Dokumentation: def-Funktionen können mit Docstrings versehen werden, was die Lesbarkeit und Wartbarkeit des Codes verbessert.
- Debugging: def-Funktionen können mit dem Debugger leichter getestet und debuggt werden.
