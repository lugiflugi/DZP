#### Was ist der Unterschied zwischen Matplotlib und Seaborn?

Seaborn und Matplotlib sind beide Python-Bibliotheken, die für die Datenvisualisierung verwendet werden können, aber es gibt einige Unterschiede zwischen ihnen.

Matplotlib ist eine grundlegende Bibliothek, die eine breite Palette von Funktionen zur Erstellung von Diagrammen bietet. Es bietet viele grundlegende Plot-Typen wie Linien-, Balken-, Scatter-, Kontur- und 3D-Plots. Matplotlib ist auch sehr anpassbar und bietet viele Optionen zur Feinabstimmung von Details wie Achsenbeschriftungen, Linienstilen und Farben.

Seaborn hingegen ist eine auf Matplotlib aufbauende Bibliothek, die speziell für die Erstellung von statistischen Plots optimiert ist. Es bietet eine einfachere Syntax und ermöglicht es, komplexe Plot-Typen mit wenigen Codezeilen zu erstellen. Es enthält auch einige fortgeschrittene Plot-Typen wie Heatmaps, Jointplots und Facetgrids, die in Matplotlib nicht verfügbar sind. Seaborn bietet auch eine Vielzahl von Farbpaletten und Stilen, die das Erstellen von ästhetisch ansprechenden Plots erleichtern.

Zusammenfassend lässt sich sagen, dass Matplotlib eine sehr mächtige und flexible Bibliothek ist, die für eine breite Palette von Anwendungen geeignet ist. Seaborn ist eine auf Matplotlib aufbauende Bibliothek, die speziell für statistische Plots optimiert ist und eine einfachere Syntax und eine Vielzahl von fortgeschrittenen Plot-Typen bietet.

### Figure-level vs. axes-level functions

In Seaborn gibt es zwei Arten von Funktionen für die Erstellung von Grafiken: Figure-level-Funktionen und Axes-level-Funktionen.

Axes-level-Funktionen sind Funktionen, die direkt auf einer einzelnen Zeichenfläche (Axes) angewendet werden, um eine bestimmte Grafik zu erstellen. Zum Beispiel kann man mit der Funktion sns.scatterplot() einen Scatterplot direkt auf einer einzelnen Axes erstellen. Mit Axes-level-Funktionen kann man die Eigenschaften der einzelnen Axes, wie z.B. Achsenbeschriftungen und Skalen, individuell anpassen.

Figure-level-Funktionen sind hingegen Funktionen, die auf der gesamten Figure angewendet werden, um mehrere Axes gleichzeitig zu erstellen und zu organisieren. Zum Beispiel kann man mit der Funktion sns.relplot() mehrere Scatterplots auf einer gemeinsamen Figure erstellen und dabei die Daten nach verschiedenen Variablen aufteilen oder gruppieren. Figure-level-Funktionen bieten eine höhere Abstraktionsebene als Axes-level-Funktionen und können daher eine schnellere und einfachere Möglichkeit bieten, komplexe Visualisierungen zu erstellen.

In der folgenden Abbildung sehen wir einige Beispiele für Figure-level-Funktionen und Axes-level-Funktionen. Die der Funktionen ist nicht abschliessend, sondern beinhaltet ein paar Beispiele, die wir in diesem Notebook verwenden werden.

- relplot() dient zur Erstellung von Beziehungsplots, die die Beziehung zwischen zwei Variablen darstellen.
- displot() dient zur Erstellung von Verteilungsplots, die die Verteilung einer einzelnen Variablen darstellen.
- catplot() dient zur Erstellung von Kategorienplots, die Daten nach einer oder mehreren Kategorien aufteilen und die Beziehungen zwischen den Variablen in den einzelnen Kategorien darstellen.

### Seaborn Catplot
Die Funktion catplot() ist eine sogenannte "figure level function" in der Seaborn-Bibliothek. 

Eine Figure level function wie catplot() kann für die Erstellung von komplexeren Diagrammen nützlich sein, bei denen mehrere Axes-Objekte benötigt werden, um verschiedene Aspekte der Daten darzustellen. Mit catplot() können Sie beispielsweise verschiedene Teildiagramme für jede Kategorie der Daten erstellen, indem Sie das Argument row oder col verwenden.

Ein typisches Anwendungsbeispiel für catplot() besteht darin, die Beziehung zwischen einer kategorischen Variablen und einer numerischen Variablen zu visualisieren. 

Es ist wichtig zu beachten, dass catplot() auf Matplotlib aufbaut und im Wesentlichen ein Wrapper für eine Kombination aus Axes level functions ist, die auf der zugrunde liegenden Axes-Struktur von Matplotlib basieren. Durch die Verwendung von catplot() können Sie jedoch komplexe Diagramme mit einer einfacheren Syntax erstellen und müssen sich nicht mit den Details der Axes-Struktur von Matplotlib befassen.

Werfen wir einen Blick auf einige der wichtigsten Argumente der Funktion catplot():

- `data=` stellt die zu plottenden Daten über einen Pandas DataFrame bereit
- `x=` und `y=` liefern die Variablen, die auf der x- bzw. y-Achse aufgetragen werden sollen
- `hue=` fügt eine zusätzliche Variable zum Plotten über eine Farbzuordnung hinzu

Darüber hinaus bietet die Funktion einige zusätzliche Parameter, die nur in der Funktion catplot() verfügbar sind.

- `kind=` bestimmt, welche Art von Diagramm erstellt werden soll. Standardmässig wird ein Stripplot mit dem Schlüsselwortargument „strip“ erstellt.
- `row=` ermöglicht es Ihnen, Ihren Datensatz in zusätzliche Zeilen von Visualisierungen aufzuteilen
- `col=` ermöglicht es Ihnen, Ihren Datensatz in zusätzliche Spalten mit Visualisierungen aufzuteilen
- `height =` und `aspect =` steuern die Grösse der Datenvisualisierung

|Categorical Scatter Plot       |	Strip Plot  | kind='strip'	    | sns.stripplot()       | 
|Categorical Scatter Plot       |	Swarm Plot  | kind='swarm'	    | sns.swarmplot()       | 
|Categorical Distribution Plot  |	Box Plot	| kind='box'	    | sns.boxplot()         |

|Categorical Estimate Plot      |	Count Plot	| kind='count'	    | sns.countplot()       | 

Schauen wir uns nun einige Beispiele an, um die Funktionsweise von catplot() zu demonstrieren.

Die Beispiel-Daten stammen aus dem Seaborn-Datensatz "tips".

Es werden alle Plot-Typen von catplot() verwendet, um verschiedene Diagramme mit dem gleichen Datensatz zu erstellen.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Laden des Datensatzes
tips = sns.load_dataset("tips")

# Setup des Plots
fig, axs = plt.subplots(ncols=3, nrows=3, figsize=(15, 10))

# Erstellen der Plots
sns.stripplot(x="day", y="total_bill", data=tips, ax=axs[0][0])
sns.swarmplot(x="day", y="total_bill", data=tips, ax=axs[0][1])
sns.boxplot(x="day", y="total_bill", data=tips, ax=axs[1][0])
sns.violinplot(x="day", y="total_bill", data=tips, ax=axs[1][1])
sns.boxenplot(x="day", y="total_bill", data=tips, ax=axs[1][2])
sns.pointplot(x="day", y="total_bill", data=tips, ax=axs[2][0])
sns.barplot(x="day", y="total_bill", data=tips, ax=axs[2][1])
sns.countplot(x="day", data=tips, ax=axs[2][2])

# Definieren der Plot Titel und  Achsenbeschriftungen
axs[0][0].set_title("Strip Plot")
axs[0][0].set_xlabel("Day of the Week")
axs[0][0].set_ylabel("Total Bill ($)")

axs[0][1].set_title("Swarm Plot")
axs[0][1].set_xlabel("Day of the Week")
axs[0][1].set_ylabel("Total Bill ($)")

axs[1][0].set_title("Box Plot")
axs[1][0].set_xlabel("Day of the Week")
axs[1][0].set_ylabel("Total Bill ($)")

axs[1][1].set_title("Violin Plot")
axs[1][1].set_xlabel("Day of the Week")
axs[1][1].set_ylabel("Total Bill ($)")

axs[1][2].set_title("Boxen Plot")
axs[1][2].set_xlabel("Day of the Week")
axs[1][2].set_ylabel("Total Bill ($)")

axs[2][0].set_title("Point Plot")
axs[2][0].set_xlabel("Day of the Week")
axs[2][0].set_ylabel("Total Bill ($)")

axs[2][1].set_title("Bar Plot")
axs[2][1].set_xlabel("Day of the Week")
axs[2][1].set_ylabel("Total Bill ($)")

axs[2][2].set_title("Count Plot")
axs[2][2].set_xlabel("Day of the Week")
axs[2][2].set_ylabel("Count")

# Setzen der Abstände zwischen den Plots
plt.subplots_adjust(hspace=0.5)

# Anzeigen des Plots
plt.show()  # wird nicht benötigt, da der Plot im Jupiter Notebook auch ohne plt.show() angezeigt wird
````

#### Erklärungen zu den Box-Plot und Boxen-Plots




Der Boxplot zeigt den Median als Mittellinie (50. Perzentil), dann das 25. und 75. Perzentil als Boxgrenzen. Dann werden die Ausreissergrenzen berechnet. Q3 ist das 3. Quartil oder 75. Perzentil der Daten (75 % der Daten liegen unter diesem Wert). Ausreisser ausserhalb der Ausreisser-Whisker werden als deutliche Punkte dargestellt.

Boxenplots zeigen die Verteilung anders und eignen sich besser für grössere Datensätze. Klassische Boxplots können zu viele Ausreisser haben und zeigen nicht so viele Informationen über die Verteilung. Boxenplots beginnen mit dem Median (Q2, 50. Perzentil) als Mittellinie. Jede nachfolgende Ebene nach aussen enthält die Hälfte der verbleibenden Daten. Die ersten beiden Abschnitte ausserhalb der Mittellinie enthalten also 50 % der Daten. Danach enthalten die nächsten beiden Abschnitte 25 % der Daten. Dies setzt sich fort, bis wir auf der Ausreisserebene sind. Jede Stufe ist heller schattiert. Es gibt 4 Methoden zur Berechnung von Ausreissern (Seaborn verfügbar). Die Standardeinstellung ist, am Ende etwa 5-8 Ausreisser in jedem Ende zu haben.

#### Ändern von Titeln und Achsenbeschriftungen in Seaborn Catplot

Das Hinzufügen von Titeln und beschreibenden Achsenbeschriftungen ist eine grossartige Möglichkeit, Ihre Datenvisualisierung kommunikativer zu gestalten. 

In vielen Fällen möchten die Leser genau wissen, was ein Datenpunkt und eine Grafik darstellen. Aus diesem Grund ist es wichtig zu verstehen, wie diese in Seaborn angepasst werden.

#### Hinzufügen von Titeln zu Seaborn Catplot

Um einem Seaborn-Catplot() einen Titel hinzuzufügen, können wir die in Matplotlib verfügbare Methode fig.suptitle() verwenden. 


````python
# Adding Titles to Seaborn Catplot

# We assigned the catplot to a variable, cat
cat = sns.catplot(
    data=df, x="island", y="bill_length_mm", kind="violin", row="sex", col="species"
)
# We then adjusted the top margin using fig.subplots_adjust()
cat.figure.subplots_adjust(top=0.92)
# Then, we passed in a suptitle() onto the figure object
cat.figure.suptitle("Comparing Penguins")
````

#### Hinzufügen von Titeln zu den Spalten und Zeilen in Seaborn Catplot

Seaborn bietet eine Reihe von Möglichkeiten, um die Titel der Spalten und Zeilen zu formatieren.

````python
# Adding Row and Column Titles

# We assigned the catplot to a variable, cat
cat = sns.catplot(
    data=df, x="island", y="bill_length_mm", kind="violin", row="sex", col="species"
)
# We then adjusted the top margin using fig.subplots_adjust()
cat.figure.subplots_adjust(top=0.92)
# Then, we passed in a suptitle() onto the figure object
cat.figure.suptitle("Comparing Penguins")
# We then added row and column titles using set_titles()
cat.set_titles(
    row_template="Penguin Sex: {row_name}", col_template="Penguin Species: {col_name}"
)
````

Im obigen Codeblock haben wir die Methode `.set_titles()` verwendet, die für FacetGrid-Objekte verfügbar ist. Die Methode ermöglicht es die Verwendung der Parameter `row_template=` und `col_template=`, die  den Zugriff auf die Variablen `col_name` und `row_name` in f-string-ähnlicher Formatierung ermöglichen.

#### Hinzufügen von Achsenbeschriftungen zu Seaborn Catplot

Standardmässig verwendet Seaborn die Spaltenbeschriftungen aus den Dataframes als Achsenbeschriftungen in der Visualisierung. In vielen Fällen ist dies jedoch kein sehr aussagekräftiger Titel. Da die Funktion catplot() ein FacetGrid-Objekt zurückgibt, können wir Hilfsmethoden verwenden, um dies zu lösen

````python
# Modifying Axis Labels in Seaborn

# We assigned the catplot to a variable, cat
cat = sns.catplot(
    data=df, x="island", y="bill_length_mm", kind="violin", row="sex", col="species"
)
# We then adjusted the top margin using fig.subplots_adjust()
cat.figure.subplots_adjust(top=0.92)
# Then, we passed in a suptitle() onto the figure object
cat.figure.suptitle("Comparing Penguins")
# We then added row and column titles using set_titles()
cat.set_titles(
    row_template="Penguin Sex: {row_name}", col_template="Penguin Species: {col_name}"
)
# We then modified the x and y axis labels using set_xlabels() and set_ylabels()
cat.set_ylabels("Bill Length (mm)")
cat.set_xlabels("Island Name")
````

#### Ändern der Grösse von Seaborn Catplots


Da die catplot()-Funktion von Seaborn ein FacetGrid-Objekt zurückgibt, können wir die Grösse des zurückgegebenen figure-Objekts leicht ändern. 
Dazu können wir die beiden folgenden Parameter verwenden:

- `height`: Höhe des Plots in Zoll
- `aspect`: Seitenverhältnis, sprich Breite (`height` * `aspect`) des Plots in Zoll

````python
# Beispiel ohne Anpassung der Plot-Grösse
# per defult ist die height=5 und die aspect=1
cat = sns.catplot(data=df, x="island", y="bill_length_mm", kind="violin")
````

````python
# Beispiel mit Anpassung der Plot-Grösse
cat = sns.catplot(
    data=df, x="island", y="bill_length_mm", kind="violin", height=4, aspect=1.6
)
````

### Seaborn Relplot

In diesem Notebook werden wir uns unter Anderem mit der Funktion sns.relplot() beschäftigen. Diese Funktion dient zur Erstellung von Beziehungsplots, die die Beziehung zwischen zwei Variablen darstellen. Die Funktion sns.relplot() ist eine Figure-level-Funktion, die auf der gesamten Figure angewendet wird, um mehrere Axes gleichzeitig zu erstellen und zu organisieren. Die Funktion sns.relplot() ist eine sehr flexible Funktion, die es uns ermöglicht, die Daten nach verschiedenen Variablen aufzuteilen oder zu gruppieren. Es ist wichtig zu beachten, dass die Funktion sns.relplot() die Axes-level-Funktion sns.scatterplot() verwendet, um die einzelnen Scatterplots zu erstellen.
Die relplot() baut auf Matplotlib auf und ist im Wesentlichen ein Wrapper für eine Kombination aus Axes level functions, die auf der zugrunde liegenden Axes-Struktur von Matplotlib basieren. Durch die Verwendung von relplot() können Sie jedoch komplexe Diagramme mit einer einfacheren Syntax erstellen und müssen sich nicht mit den Details der Axes-Struktur von Matplotlib befassen.

Folgende Parametern der Funktion sns.relplot() sind besonders wichtig:

- `data=`: Die Daten, die für die Erstellung der Grafik verwendet werden sollen.
- `x=`: Die Variable, die auf der x-Achse dargestellt werden soll.
- `y=`: Die Variable, die auf der y-Achse dargestellt werden soll.
- `hue=`: Die Variable, nach der die Daten gruppiert werden sollen. Die Daten werden nach den verschiedenen Ausprägungen der Variable gruppiert und jede Gruppe wird in einer anderen Farbe dargestellt.
- `row=`/`col=`: Die Variable, nach der die Daten in einzelne Zeilen/Spalten aufgeteilt werden sollen. Jede Zeile/Spalte enthält die Daten einer anderen Ausprägung der Variable.
- `kind=`: Die Art der Grafik, die erstellt werden soll. Die möglichen Werte sind "scatterplot" (Streudiagramm) und "lineplot" (Liniendiagramm).
- `size=`: Die Variable, die die Grösse der einzelnen Punkte bestimmt. Die Grösse der Punkte wird anhand der Werte der Variable skaliert.
- `style=`: Die Variable, die die Form der einzelnen Punkte bestimmt. Die Form der Punkte wird anhand der Werte der Variable bestimmt.
- `palette=`: Die Farbpalette, die für die Darstellung der Daten verwendet werden soll. 
- `height=`: Die Höhe der einzelnen Axes.
- `aspect=`: Das Seitenverhältnis der einzelnen Axes.


Die Beschreibung der weiteren Parameter finden ihr in der [Dokumentation](https://seaborn.pydata.org/generated/seaborn.relplot.html).

#### Laden von Datensets

Die nächsten Beispiel-Daten stammen aus dem Seaborn-Datensatz "flights". Dieser Datensatz enthält Informationen über die Anzahl der Passagiere, die in den Jahren 1949 bis 1960 pro Monat auf Flügen transportiert wurden.

Der Datensatz enthält die folgenden Variablen:
- `year` - Jahr, in dem die Daten aufgezeichnet wurden
- `month` - Monat, in dem die Daten aufgezeichnet wurden
- `passengers` - Anzahl der Passagiere, die in diesem Monat transportiert wurden

Es werden alle Plot-Typen von relplot() verwendet, um verschiedene Diagramme mit dem gleichen Datensatz zu erstellen.

Die Beispiel-Daten stammen aus dem Seaborn-Datensatz "tips"

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Laden des Datensatzes
flights = sns.load_dataset("flights")
flights
````

````python
# Setup des Plots
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(15, 10))

# Erstellen der Plots
sns.scatterplot(x="year", y="passengers", data=flights, ax=axs[0][0])
sns.lineplot(x="year", y="passengers", data=flights, ax=axs[0][1])

sns.scatterplot(x="year", y="passengers", data=flights, hue="month", ax=axs[1][0])
sns.lineplot(x="year", y="passengers", data=flights, hue="month", ax=axs[1][1])

# Definieren der Plot Titel und  Achsenbeschriftungen
axs[0][0].set_title("Scatter Plot")
axs[0][0].set_xlabel("Jahr")
axs[0][0].set_ylabel("Passagiere")

axs[0][1].set_title("Line Plot")
axs[0][1].set_xlabel("Jahr")
axs[0][1].set_ylabel("Passagiere")

axs[1][0].set_title("Scatter Plot")
axs[1][0].set_xlabel("Jahr")
axs[1][0].set_ylabel("Passagiere")
axs[1][0].legend(title="Monat")

axs[1][1].set_title("Line Plot")
axs[1][1].set_xlabel("Jahr")
axs[1][1].set_ylabel("Passagiere")
axs[1][1].legend(title="Monat")

# Setzen der Abstände zwischen den Plots
plt.subplots_adjust(hspace=0.5)

# Anzeigen des Plots
plt.show()  # wird nicht benötigt, da der Plot im Jupiter Notebook auch ohne plt.show() angezeigt wird
````

####  Relplot - Basic Scatterplot (default)

Bei der statistischen Analyse geht es darum, zu verstehen, wie Variablen in einem Datensatz zueinander in Beziehung stehen und wie diese Beziehungen von anderen Variablen abhängen.

Es gibt verschiedene Möglichkeiten, in Seaborn ein Streudiagramm zu zeichnen. Die einfachste Funktion, die verwendet werden sollte, wenn beide Variablen numerisch sind, ist die Funktion „Scatterplot()“

Schauen wir uns an ob es einen Zusammenhang zwischen der Schnabellänge und der Schnabeltiefe gibt.

````python
sns.relplot(x="bill_length_mm", y="bill_depth_mm", data=penguins)
````

#### Relplot - Scatterplot mit Farbgruppierung, Markergrösse und Markerform

Schauen wir uns an ob es einen Zusammenhang zwischen der Schnabellänge und der Schnabeltiefe gibt und welchen Einfluss z.B. die Art des Tieres hat.

Während die Punkte zweidimensional dargestellt werden, kann dem Diagramm eine weitere Dimension hinzugefügt werden, indem die Punkte entsprechend einer dritten Variablen eingefärbt werden. In Seaborn wird dies als Verwendung einer „Farbtonsemantik“ bezeichnet, da die Farbe des Punktes an Bedeutung gewinnt:

````python
# sns.relplot(
#     x="bill_length_mm", y="bill_depth_mm", data=penguins, hue="species", kind="scatter"
# )
# sns.relplot(
#     x="bill_length_mm", y="bill_depth_mm", data=penguins, size="species", kind="scatter"
# )
# sns.relplot(
#     x="bill_length_mm",
#     y="bill_depth_mm",
#     data=penguins,
#     style="species",
#     kind="scatter",
# )

sns.relplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    data=penguins,
    hue="species",
    size="species",
    style="species",
    kind="scatter",
)
````

Es ist auch möglich, vier Variablen darzustellen, indem man den Farbton und Stil jedes Punkts unabhängig ändert. 

Dies sollte jedoch sorgfältig erfolgen, da das Auge viel weniger empfindlich auf die Form als auf die Farbe reagiert:

````python
sns.relplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    data=penguins,
    hue="species",
    style="sex",
    kind="scatter",
)
````

Statt den Stil des Markers zu ändern kann auch die Grösse des Markers angepasst werden. 

````python
sns.relplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    data=penguins,
    hue="species",
    size="body_mass_g",
    kind="scatter",
)
````

In den obigen Beispielen war die Farbtonsemantik kategorisch, sodass die standardmässige qualitative Farbpalette angewendet wurde. 

Wenn die Farbtonsemantik numerisch ist (insbesondere, wenn sie in Float umgewandelt werden kann), wechselt die Standardfärbung zu einer sequentiellen Palette:

Um dieses Beispiel zu zeigen wechsle ich auf den Datensatz "tips".

````python
sns.relplot(data=tips, x="total_bill", y="tip", hue="size")
````

Die Farbpalette kann mit dem Parameter `palette=` angepasst werden. Es gibt verschiedene Möglichkeiten die Farbpalette zu definieren. Eine dieser Möglichkeiten ist die Verwendung von [cubehelix_palette()](https://seaborn.pydata.org/generated/seaborn.cubehelix_palette.html#seaborn.cubehelix_palette).

Eine weitere Möglichkeit ist die Verwendung von [Farbpaletten](https://medium.com/@morganjonesartist/color-guide-to-seaborn-palettes-da849406d44f).

````python
cubehelix_palette = sns.cubehelix_palette(rot=-0.5, light=0.75)
sns.relplot(data=tips, x="total_bill", y="tip", hue="size", palette=cubehelix_palette)
````

Eine weitere Möglichkeit einen Datenpunkt hervorzuheben ist die Verwendung von `size=`. Bei numerischen Variablen wird die Grösse der Punkte anhand der Werte der Variable skaliert.

````python
sns.relplot(data=tips, x="total_bill", y="tip", size="size")
````

Das Argument `sizes=` kann verwendet werden, um die Grösse der Punkte zu ändern. 

Es kann entweder eine numerische Variable sein, die die Grösse der einzelnen Punkte bestimmt, oder ein Tupel, das die minimalen und maximalen Grössen der Punkte angibt.

````python
sns.relplot(data=tips, x="total_bill", y="tip", size="size", sizes=(15, 200))
````

#### Relplot - Lineplot Messungen über die Zeit mit einem Datenpunkt pro Messung

Streudiagramme sind sehr effektiv, es gibt jedoch keine allgemein optimale Visualisierungsart. Stattdessen sollte die visuelle Darstellung an die Besonderheiten des Datensatzes und an die Frage angepasst werden, die ihr mit dem Diagramm beantworten möchten.

Bei einigen Datensätzen möchte man möglicherweise Änderungen in einer Variablen als Funktion der Zeit oder einer ähnlich kontinuierlichen Variablen darstellen. Hier kann ein Liniendiagramm verwendet werden.

````python
dowjones = sns.load_dataset("dowjones")
sns.relplot(data=dowjones, x="Date", y="Price", kind="line")
````

#### Relplot - Lineplot mit mehreren Messungen pro auf der x-Achse

Komplexere Datensätze verfügen über mehrere Messungen für denselben Wert der x-Variablen

Das Standardverhalten in Seaborn ist, die mehreren Messungen bei jedem x-Wert zu aggregieren, indem der Mittelwert und das 95 %-Konfidenzintervall um den Mittelwert aufgetragen werden:

````python
fmri = sns.load_dataset("fmri")
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line")
````

````python
sns.relplot(
    data=fmri, kind="line", x="timepoint", y="signal", hue="event", style="event"
)
````

Definiert man den style-Parameter als `style="event"` werden die Muster der Linien verändert.

````python
sns.relplot(
    data=fmri,
    kind="line",
    x="timepoint",
    y="signal",
    hue="region",
    style="event",
)
````

Man kann auf die Darstellung der Linien Einfluss nehmen, indem ihr die einzelnen Messpunkte heraushebt mit dem Parameter `markers=`.

````python
sns.relplot(
    data=fmri,
    kind="line",
    x="timepoint",
    y="signal",
    hue="region",
    style="event",
    dashes=False,
    markers=True,
)
````

### Seaborn Heatmap
Mit einer Heatmap können wir Beziehungen zwischen drei Variablen auf einer 2D-Ebene visualisieren. Diese Beziehungen können kompliziert sein, deshalb wird Farbe verwendet. Es kann wichtige Aspekte der Beziehung hervorheben und sie verständlicher machen.

Die Heatmaps hat seine Grenzen. Die x-Variable y-Variable müssen diskret oder kategorisch sein. Oder, wenn sie kontinuierlich sind, müssen wir sie in Gruppen einteilen können. Andererseits muss Werte-Variable eine kontinuierliche Variable sein. 

#### Heatmap - Basic Heatmap

````python
# heatmap mit tips dataset machen
grouped_tips = tips.groupby(["day", "time"])["total_bill"].mean().reset_index()
grouped_tips = grouped_tips.pivot(index="day", columns="time", values="total_bill")

# sns.heatmap(grouped_tips, cmap='YlOrRd')
sns.heatmap(grouped_tips, cmap="YlOrRd", annot=True, fmt=".1f", linewidths=0.5)
````

Für das nächste Beispiel verwenden wir die globale Temperaturdaten von der NASA  [link](https://data.giss.nasa.gov/gistemp/).
Die Daten sind seit 1880 verfügbar und werden als monatliche Durchschnittstemperaturen in Grad Celsius angegeben.

````python
df = pd.read_csv("monthly_csv.csv")

# Nur die Datensätze aus der Quelle GISTEMP verwenden
df = df[df.Source == "GISTEMP"]

# Date in datetime umwandeln
df["Date"] = pd.to_datetime(df["Date"])

# Neue Spalten mit dem Jahr und Monat erstellen
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month

df.head()
````

````python
# Erstelle ein Matrix mit den Jahren als Spalten und den Monaten als Zeilen und den Temperaturen als Werte
df_pivot = df.pivot(index="Month", columns="Year", values="Mean")
df_pivot
````

````python
sns.heatmap(df_pivot, cmap="coolwarm", xticklabels=10)
````

Darstellen von Korrelationen mit einer Heatmap. Für das nächste Beispiel verwenden wir den "pengiun"-Datensatz.

````python
# Korrelationsmatrix berechnen
corr_matrix = penguins[
    ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
].corr()

# Heatmap erstellen
sns.heatmap(corr_matrix, annot=True, cmap="bwr")
````

#### Scaling Plots

Die `set_context()` Funktion ermöglicht es Ihnen, die Grösse der Visualisierung zu ändern. Es gibt vier voreingestellte Kontexte: `paper`, `notebook`, `talk` und `poster`. Der `notebook`-Kontext ist der Standard.

````python
sns.set_style("darkgrid")
# Smallest context:
sns.set_context("paper")
sns.stripplot(x="day", y="total_bill", data=tips)
````

````python
sns.set_context("poster")
sns.stripplot(x="day", y="total_bill", data=tips)
````

````python
sns.set_context("poster", rc={"grid.linewidth": 0.8}, font_scale=0.8)
sns.stripplot(x="day", y="total_bill", data=tips)
````
